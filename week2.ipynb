# This task involves the pandas library to manipulate the data

# Load a CSV file into pandas DataFrame and Performing operations like filtering data based on conditions,handling missing values and calculating summary statistics...

```python
import pandas as pd
```

```python
# importing the csv file
data=pd.read_csv('01.Data Cleaning and Preprocessing.csv')
data
```

```python
df=pd.DataFrame(data)

df
```

# Handling Missing Values...

```python
# Checking Null values..
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
# Filling null values..
df1=df.fillna(value=0)
df1
```

```python
# Filling Null values with Previous value...
df2=df.fillna(method='pad')
df2
```

```python
# Filling Null values with the next value..
df3=df.fillna(method='bfill')
df3
```

```python
df.shape
```

```python
df4=df.dropna()
df4.shape
```

# Filtering Data basedd on conditions...

```python
# using loc function..
df.loc[df['BlowFlow']>100.00]
```

```python
df.loc[df['ChipRate']<60.00]
```

```python
# using and operator..
df.loc[(df['BlowFlow']>100.00) & (df['ChipRate']<60.00)]
```

# Calculating Summary Statistics...

```python
# using count() function..
df.count()
```

```python
# Using value_counts() on specific column..
df['ChipRate'].value_counts()
```

```python
# Using min() function on a specific row
df['ChipRate'].min()
```

```python
# Using max() function on a specific row..
df['ChipRate'].max()
```

```python
# mean() gives the average value of a specific row..
df['ChipRate'].mean()
```

```python
# median() gives the median value of a specific row..
df['ChipRate'].median()
```

```python
# mode() gives the most repeated value of a specific row..
df['ChipRate'].mode()
```
